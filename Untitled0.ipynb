```python
import matplotlib.pyplot as plt
import numpy as np

def jdro1(x):
  #if(x>=0 & x<0):
  #  return 1
  #else: return 0
  return (x>=0)*(x<1)

def jdro2(x):
    return (x>=-0.5)*(x<0.5)


def jdro3(x):
    # 1 - |x| dla |x| <= 1, inaczej 0
    y = np.zeros_like(x)
    mask = np.abs(x) <= 1
    y[mask] = 1 - np.abs(x[mask])
    return y

def jdro4(x):
    # sin(x)/x z uwzgldnieniem x = 0
    y = np.ones_like(x)
    mask = x != 0
    y[mask] = np.sin(x[mask]) / x[mask]
    return y

x1 = np.linspace(-1, 2, 100)
y1 = jdro1(x1)
x2 = np.linspace(-1, 2, 100)
y2 = jdro2(x2)
x3 = np.linspace(-1, 2, 100)
y3 = jdro3(x1)
x4 = np.linspace(-1, 2, 100)
y4 = jdro4(x2)


plt.figure(figsize=(8,8))
plt.plot(x1, y1, label='jdro1')
plt.plot(x2, y2, label='jdro2')
plt.plot(x3, y3, label='jdro3')
plt.plot(x4, y4, label='jdro4')
plt.legend(fontsize=11)
plt.show()
```
